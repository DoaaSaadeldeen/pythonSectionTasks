```python
import networkx as nx
%matplotlib inline
```

```python
G=nx.Graph()
```

```python
G.add_nodes_from(['a','b','c','d','e','f'])
G.add_edges_from([('a','b'),('a','c'),('b','d'),('b','e'),('c','f')])
nx.draw(G,with_labels=True)
```

```python
help(nx.Graph())
```

```python
#to add edge or edges
G.add_edge('a','b')
#or a list of edges
edges=[('a','b'),('b','c'),('c','d'),('d','e'),('e','a')]
G.add_edges_from(edges)
```

```python
#to draw the graph
nx.draw(G,with_labels=True,node_color='green',font_color='white',node_size=1600,font_size=16)
```
